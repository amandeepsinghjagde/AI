```python
import numpy as np;
import pandas as pd;
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
data
```

```python
#A
from keras.models import Sequential
from keras.layers import Dense
```

```python
df_label = data['Outcome']
df_features = data.drop('Outcome', 1)
df_features.replace('?', -99999, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataSetA = data.iloc[:,0:3]
```

```python
x_train, x_test, y_train, y_test = train_test_split(dataSetA, df_label, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Dense(8, input_dim=3, activation='relu'))
model.add(Dense(2, activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['accuracy'])
model.fit(x_train,y_train, epochs=100, batch_size=10, validation_data=(x_test, y_test))
```
